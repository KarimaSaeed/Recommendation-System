```python
from sklearn.feature_extraction.text import TfidfVectorizer

tf_idf = TfidfVectorizer(dtype= np.float32)

tfidf_matrix = tf_idf.fit_transform(movies['genres'])
```

```python
# Take unique movie ids from movies table
movie_idx = {val : idx for idx, val in enumerate(movies['movieId'])}
# Then map these values to moviedId in rating table
rating['movie_idx'] = rating['movieId'].map(movie_idx)
```

```python
rating
```

```python
from scipy.sparse import csr_matrix

user_movie_matrix = csr_matrix( (rating['rating'], (rating['userId'], rating['movie_idx'])), dtype= np.float32)
```

```python
import streamlit as st
```

```python
%%writefile recommendation_app.py

import streamlit as st
import pandas as pd
import joblib
import numpy as np

tfidf_matrix = joblib.load('tfidf_matrix.pkl')
user_profile = joblib.load('user_profile.pkl')
movies = pd.read_parquet('movies.parquet')

user_id = st.number_input('User ID', step= 1)
top_n = st.slider('Top N', step= 1)

user_vector = tfidf_matrix @ user_profile.toarray()[user_id]

score_idx = np.argsort(user_vector)[::-1][:top_n]

st.subheader('Top Recommended Movies')

for idx, mov in enumerate(movies['title'].iloc[score_idx]):

    st.write(f'{idx + 1} : {mov}')
```
